```python
df_og = pd.read_csv('../data/train.csv')
df_og.shape
```

```python
# ELIMINO LOS NANS PARA PODER OPERAR NUMERICAMENTE
df = df_og.dropna()
df = df.reset_index()
df.shape
```

```python
# Comenzaremos intentando determinar que variables estan mas estrechamente relacionadas, tanto proporcional
# como inversamente proporcional.
# ELIMINAMOS:
# 1) lat y lng
# 2) id
# 3) index

y = df['precio'].values
y = y.reshape(len(y),1)

column_list = list(df)
column_list_numbers = []

for i in range(0,len(column_list)):
    if isinstance(df[column_list[i]][0], numbers.Number) and column_list[i] != 'id' and column_list[i] != 'index' and column_list[i] != 'idzona':
        column_list_numbers.append(column_list[i])
column_list_numbers
```

```python
# TODOS CONTRA PRECIO CON KFOLD = K
K = 5

errores_promedios = pd.DataFrame(column_list_numbers, columns = ['Caracteristica'])
error_promedio_actual_log = []
error_promedio_actual = []
error_promedio_actual_r2 = []
for i in range(0,len(column_list_numbers)):
    x = df[column_list_numbers[i]].values
    x = x.reshape(len(x),1)
    kfxv = KfoldXvalidation(x,y,K,1)
    promedio_actual_log = sum(kfxv.rmsle_por_fold()) / K
    promedio_actual = sum(kfxv.rmse_por_fold()) / K
    promedio_actual_r2 = sum(kfxv.r_squared()) / K
    error_promedio_actual_log.append(round(float(promedio_actual_log),2))
    error_promedio_actual.append(int(promedio_actual / 1000))
    error_promedio_actual_r2.append(round(float(promedio_actual_r2),2))
```

```python
errores_promedios['RMSLE'] = error_promedio_actual_log
errores_promedios['RMSE (x 1000)'] = error_promedio_actual
errores_promedios['R^2'] = error_promedio_actual_r2
errores_promedios
```

```python
from sklearn import linear_model

for nombre in column_list_numbers:
    colum = df[nombre].values
    colum = colum.reshape(len(colum), 1)
    precio = df["precio"].values
    precio = precio.reshape(len(precio), 1)
    
    punto_de_corte = (len(colum)*8)//10
    colum_train = colum[0:punto_de_corte]
    precio_train = precio[0:punto_de_corte]
    colum_test = colum[punto_de_corte:]
    precio_test = precio[punto_de_corte:]
    
    clf = linear_model.LinearRegression()
    clf.fit(colum_train, precio_train)
    prediction = clf.predict(colum_test)
    prediction = prediction.reshape(len(prediction))

    colum_test = colum_test.reshape(len(colum_test))
    precio_test = precio_test.reshape(len(precio_test))
    d = {nombre : colum_test, "precio" : precio_test, "prediccion" : prediction}
    test_scatter = pd.DataFrame(data = d)
    sns.scatterplot(data = test_scatter, x = nombre, y = "precio")
    
    #d["prediccion"] = prediction
    sns.lineplot(data = test_scatter, x = nombre, y = "prediccion", color = "red")
    plt.savefig("analisis_preliminar_"+nombre+"sklearn.png")
    plt.show()
    
```
