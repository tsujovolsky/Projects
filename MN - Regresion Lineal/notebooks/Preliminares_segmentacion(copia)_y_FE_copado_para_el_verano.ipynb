```python
# Comenzaremos intentando determinar que variables estan mas estrechamente relacionadas, tanto proporcional
# como inversamente proporcional.

y = df['precio'].values
y = y.reshape(len(y),1)

column_list = list(df)
column_list_numbers = []

for i in range(0,len(column_list)):
    if isinstance(df[column_list[i]][0], numbers.Number) and column_list[i] != 'lat' and column_list[i] != 'lng':
        column_list_numbers.append(column_list[i])
column_list_numbers
```

```python
# TODOS CONTRA PRECIO CON KFOLD = K
K = 5

errores_promedios = pd.DataFrame(column_list_numbers, columns = ['Caracteristica'])
errorLog_promedio_actual = []
error_promedio_actual = []
for i in range(0,len(column_list_numbers)):
    x = df[column_list_numbers[i]].values
    x = x.reshape(len(x),1)
    kfxv = KfoldXvalidation(x,y,K, 1)
    promedio_log_actual = sum(kfxv.rmsle_por_fold()) / K
    promedio_actual = sum(kfxv.rmse_por_fold()) / K
    errorLog_promedio_actual.append(promedio_log_actual)
    error_promedio_actual.append(int(promedio_actual))
```

```python
errores_promedios['rmsle promedio'] = errorLog_promedio_actual
errores_promedios['rmse promedio'] = error_promedio_actual
errores_promedios
```
