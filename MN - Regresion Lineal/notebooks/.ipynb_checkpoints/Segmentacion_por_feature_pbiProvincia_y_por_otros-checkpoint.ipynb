```python
# CALCULOS SEGMENTANDO
lim_pbi_bajo = 300000
lim_pbi_medio = 1000000
lim_pbi_alto = 2000000

df_pbi_bajo = df[df['provincia'] < lim_pbi_bajo]
df_pbi_medio = df[df['provincia'] < lim_pbi_medio]
df_pbi_alto = df[df['provincia'] < lim_pbi_alto]
df_pbi_maximo = df[df['provincia'] >= lim_pbi_alto]

y_df_pbi_bajo = df_pbi_bajo['precio'].values
y_df_pbi_bajo = y_df_pbi_bajo.reshape(len(y_df_pbi_bajo),1)
y_df_pbi_medio = df_pbi_medio['precio'].values
y_df_pbi_medio = y_df_pbi_medio.reshape(len(y_df_pbi_medio),1)
y_df_pbi_alto = df_pbi_alto['precio'].values
y_df_pbi_alto = y_df_pbi_alto.reshape(len(y_df_pbi_alto),1)
y_df_pbi_maximo = df_pbi_maximo['precio'].values
y_df_pbi_maximo = y_df_pbi_maximo.reshape(len(y_df_pbi_maximo),1)

df_pbi_bajo = df_pbi_bajo.drop(columns='precio')
df_pbi_medio = df_pbi_medio.drop(columns='precio')
df_pbi_alto = df_pbi_alto.drop(columns='precio')
df_pbi_maximo = df_pbi_maximo.drop(columns='precio')

column_list = list(df_pbi_alto)
column_list
```

```python
#CALCULOS CON LA SEGMENTACION POR PBI DE PROVINCIAS
# PBI BAJO

K = 5

error_normal = []
error_log = []
error_r2 = []

kfxv = KfoldXvalidation(df_pbi_bajo,y_df_pbi_bajo,K,1)
promedio_rmsle = sum(kfxv.rmsle_por_fold()) / K
promedio_rmse = sum(kfxv.rmse_por_fold()) / K
promedio_r2 = sum(kfxv.r_squared()) / K

error_df_bajo = pd.DataFrame([[
    int(promedio_rmse / 1000),
    round(float(promedio_rmsle),2),
    round(float(promedio_r2),2),
]],columns=['RMSE x1000','RMSLE','R^2'])
error_df_bajo
```

```python
#CALCULOS CON LA SEGMENTACION POR PBI DE PROVINCIAS
# PBI MEDIO

K = 5

error_normal = []
error_log = []
error_r2 = []

kfxv = KfoldXvalidation(df_pbi_medio,y_df_pbi_medio,K,1)
promedio_rmsle = sum(kfxv.rmsle_por_fold()) / K
promedio_rmse = sum(kfxv.rmse_por_fold()) / K
promedio_r2 = sum(kfxv.r_squared()) / K

error_df_medio = pd.DataFrame([[
    int(promedio_rmse / 1000),
    round(float(promedio_rmsle),2),
    round(float(promedio_r2),2),
]],columns=['RMSE x1000','RMSLE','R^2'])
error_df_medio
```

```python
#CALCULOS CON LA SEGMENTACION POR PBI DE PROVINCIAS
# PBI ALTO

K = 5

error_normal = []
error_log = []
error_r2 = []

kfxv = KfoldXvalidation(df_pbi_alto,y_df_pbi_alto,K,1)
promedio_rmsle = sum(kfxv.rmsle_por_fold()) / K
promedio_rmse = sum(kfxv.rmse_por_fold()) / K
promedio_r2 = sum(kfxv.r_squared()) / K

error_df_alto = pd.DataFrame([[
    int(promedio_rmse / 1000),
    round(float(promedio_rmsle),2),
    round(float(promedio_r2),2),
]],columns=['RMSE x1000','RMSLE','R^2'])
error_df_alto
```
