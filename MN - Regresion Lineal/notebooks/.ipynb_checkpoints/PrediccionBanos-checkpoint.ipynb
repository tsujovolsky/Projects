```python
df = pd.read_csv('../data/train.csv')
df
```

```python
columns = list(df)
percent_incomplete = []
for i in columns:
    percent_incomplete.append(int(100 - round(df[i].isnull().sum() / 2400)))
df_incomplete = df.isnull().sum()
df_incomplete = pd.DataFrame(df_incomplete,columns=['Cantidad de NaNs'])
df_incomplete.insert(1,'Porcentaje completo',percent_incomplete)
print (df_incomplete)
```

```python
df = df[df['tiposdepropiedad']=='casa'].dropna()
```

```python
x = df['habitaciones'].values
print(x)
```

```python
y = df['banos'].values
```

```python
x = x.reshape(len(x),1)
y = y.reshape(len(y),1)
```

```python
x.shape
```

```python
from sklearn.linear_model import LinearRegression

"""
Reemplazar por nuestra implementación 
"""

linear_regressor1 = LinearRegression()
linear_regressor = metnum.LinearRegression()

linear_regressor.fit(x,y)
linear_regressor1.fit(x,y)
```

```python
df['prediction'] = linear_regressor.predict(x)
df['prediction1'] = linear_regressor1.predict(x)
```

```python
sns.scatterplot(data=df, x='habitaciones', y='banos')
sns.lineplot(data=df, x='metroscubiertos', y='prediction', color='red')
sns.lineplot(data=df, x='metroscubiertos', y='prediction1', color='green')
plt.figure(figsize = (1000,1000))
```
