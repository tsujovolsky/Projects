```python
# CALCULOS SEGMENTANDO
lim_pbi_bajo = 300000
lim_pbi_medio = 1000000
lim_pbi_alto = 2000000

df_pbi_bajo = df[df['provincia'] < lim_pbi_bajo]

df_pbi_medio = df[df['provincia'] >= lim_pbi_bajo]
df_pbi_medio = df_pbi_medio[df_pbi_medio['provincia'] < lim_pbi_medio]


df_pbi_alto = df[df['provincia'] >= lim_pbi_medio]
df_pbi_alto = df_pbi_alto[df_pbi_alto['provincia'] < lim_pbi_alto]

df_pbi_maximo = df[df['provincia'] >= lim_pbi_alto]
```

```python
propiedades = df_og["tipodepropiedad"].dropna().unique()
cant_propiedades = len(propiedades)
cant_propiedades
```

```python
rmsle_fold1 = []
rmsle_fold2 = []
rmsle_fold3 = []
rmsle_fold4 = []
rmsle_fold5 = []
rmse_promedio = []
r2_promedio = []
datos_totales = []
datos_sin_NaN = []
criterios = ["habitaciones", "antiguedad", "garages", "banos", "metroscubiertos", "metrostotales"]
criterios.append("precio")
i = 0
propiedades_analisis = []
df_para_analisis = df_pbi_bajo.copy() #le vamos a ir dropeando ciudades
cant_datos_de_corte = 300
for propiedad in propiedades:
    
    df = df_para_analisis[df_para_analisis["tipodepropiedad"] == propiedad]
    cant_sin_NaN = df[criterios].dropna().shape[0]
    if cant_sin_NaN > cant_datos_de_corte:
        propiedades_analisis.append(propiedad)
        datos_totales.append( df.shape[0])

        df = df[criterios].dropna()
        datos_sin_NaN.append(df.shape[0])

        y = df['precio'].values
        y = y.reshape(len(y),1)
        x = df.drop(columns = ["precio"]).values

        kfxv = kfxv = KfoldXvalidation(x, y, 5, 1)

        rmse_promedio.append(int(np.around(np.mean(kfxv.rmse_por_fold())))//1000)

        r2_promedio.append(round(float(np.mean(kfxv.r_squared())), 2))

        rmsle = kfxv.rmsle_por_fold()
        rmsle_fold1.append(rmsle[0])
        rmsle_fold2.append(rmsle[1])
        rmsle_fold3.append(rmsle[2])
        rmsle_fold4.append(rmsle[3])
        rmsle_fold5.append(rmsle[4]) 
        
        df_para_analisis.drop(df_para_analisis[df_para_analisis["tipodepropiedad"] == propiedad].index, inplace = True)
    
    
    i += 1

propiedades_analisis.append("Otros")
datos_totales.append(df_para_analisis.shape[0])
df = df_para_analisis[criterios].dropna()
datos_sin_NaN.append(df.shape[0])

y = df['precio'].values
y = y.reshape(len(y),1)
x = df.drop(columns = ["precio"]).values

kfxv = kfxv = KfoldXvalidation(x, y, 5, 1)

rmse_promedio.append(int(np.around(np.mean(kfxv.rmse_por_fold())))//1000)

r2_promedio.append(round(float(np.mean(kfxv.r_squared())), 2))

rmsle = kfxv.rmsle_por_fold()
rmsle_fold1.append(rmsle[0])
rmsle_fold2.append(rmsle[1])
rmsle_fold3.append(rmsle[2])
rmsle_fold4.append(rmsle[3])
rmsle_fold5.append(rmsle[4])
    
df_tipo_propiedad_analisis = pd.DataFrame()
df_tipo_propiedad_analisis["tipo de propiedad"] = propiedades_analisis
df_tipo_propiedad_analisis["datos totales"] = datos_totales
df_tipo_propiedad_analisis["datos sin NaN"] = datos_sin_NaN
df_tipo_propiedad_analisis["RMSE promedio"] = rmse_promedio
df_tipo_propiedad_analisis["R square promedio"] = r2_promedio
```

```python
pd.set_option('display.max_rows', df_tipo_propiedad_analisis.shape[0]+1)
print("pbi bajo")
display(df_tipo_propiedad_analisis)
print(df_tipo_propiedad_analisis.to_latex())
```
